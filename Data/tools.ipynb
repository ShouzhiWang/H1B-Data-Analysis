```python
df = pd.read_csv('/Users/shawnwang/Downloads/H1B Data Analysis/data/companies_sorted.csv')

df.head()# tabs as delimiters


```

```python
# Define size ranges for medium and large
medium_sizes = ['51 - 200', '201 - 500', '501 - 1000']
large_sizes = ['1001 - 5000', '5001 - 10000', '10001+']

# Filter medium companies and save names
df[df['size range'].isin(medium_sizes)][['name']].to_csv('Data/medium_companies.csv', index=False)

# Filter large companies and save names
df[df['size range'].isin(large_sizes)][['name']].to_csv('Data/large_companies.csv', index=False)

print("CSV files created:")
print(" - /mnt/data/medium_companies.csv")
print(" - /mnt/data/large_companies.csv")
```

```python
df = pd.read_excel('/Users/shawnwang/Downloads/H1B Data Analysis/data/natsector_M2024_dl.xlsx')

df.head()# tabs as delimiters
```

```python
filtered = df[(df['AREA'] == 99) & (df['AREA_TYPE'] == 1)]

# 3. Keep only the necessary columns
keep = [
    'AREA', 'AREA_TITLE', 'AREA_TYPE', 'PRIM_STATE',
    'NAICS', 'NAICS_TITLE',
    'A_PCT10', 'A_PCT25', 'A_MEDIAN', 'A_PCT75', 'A_PCT90',
    'ANNUAL'
]
clean = filtered[keep]

# (Optional) Rename for clarity
clean = clean.rename(columns={
    'A_PCT10': 'Annual_Pct10',
    'A_PCT25': 'Annual_Pct25',
    'A_MEDIAN': 'Annual_Median',
    'A_PCT75': 'Annual_Pct75',
    'A_PCT90': 'Annual_Pct90',
    'ANNUAL': 'Annual_Mean'
})

# 4. Export to CSV
clean.to_csv('oe_industry_wages_clean.csv', index=False)

print("Cleaned OEWS wage data written to oe_industry_wages_clean.csv")
```
